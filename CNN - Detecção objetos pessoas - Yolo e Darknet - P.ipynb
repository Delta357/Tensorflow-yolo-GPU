```python
# Imagem visualizada

import cv2
import matplotlib.pyplot as plt

def imagem(acesso):
  img = cv2.imread(acesso)
  fig = plt.gcf()
  fig.set_size_inches(18, 10)
  plt.axis("Off")
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
